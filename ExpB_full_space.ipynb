```python
device = 'cuda:3'

img = plt.imread('ExpB/data/mra.jpeg')[40:-25, 147:-206, 0] / 255

img_tensor = torch.from_numpy(img)[None, None, :, :].double().to(device) 
mask = torch.ones((1, 1, 512, 512)).double().to(device) 

img_tensor = F.pad(torch.from_numpy(img)[None, None, :, :].double().to(device), (1, 1, 1, 1))
mri_in = img_tensor[:, :, :-1, :-1]
mri = MRI(mask, device)
mri.set_h(1)
y = mri.H(mri_in)

mrs = MultiResSolver('fm', 'htv', lmbda =0, mask=mask, h_init=8, N_scales=6, range_r=512, device=device, verbose=True, toi=1e-6)
mrs.solve_fm(y)
```

```python
torch.save(mrs.sols[0].to('cpu'), 'ExpB/saved_results/mrs_full0.pt')
torch.save(mrs.sols[1].to('cpu'), 'ExpB/saved_results/mrs_full1.pt')
torch.save(mrs.sols[2].to('cpu'), 'ExpB/saved_results/mrs_full2.pt')
torch.save(mrs.sols[3].to('cpu'), 'ExpB/saved_results/mrs_full3.pt')
torch.save(mrs.sols[4].to('cpu'), 'ExpB/saved_results/mrs_full4.pt')
torch.save(mrs.sols[5].to('cpu'), 'ExpB/saved_results/mrs_full5.pt')

torch.save(mrs.costs, 'ExpB/saved_results/mrs_full_cost.pt')
torch.save(mrs.losses, 'ExpB/saved_results/mrs_full_loss.pt')
torch.save(mrs.iters, 'ExpB/saved_results/mrs_full_iters.pt')
torch.save(img_tensor, 'ExpB/saved_results/mrs_full_img.pt')
torch.save(y, 'ExpB/saved_results/mrs_full_y.pt')
```
