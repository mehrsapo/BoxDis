```python
device = 'cuda:3'

img = plt.imread('ExpB/data/mra.jpeg')[40:-25, 147:-206, 0] / 255

img_tensor = torch.from_numpy(img)[None, None, :, :].double().to(device) 
mat = scipy.io.loadmat('ExpB/masks/radial_mask.mat')
mask = torch.from_numpy(mat['mask']).double().to(device).view([1, 1, 256, 256])

img_tensor = F.pad(torch.from_numpy(img)[None, None, :, :].double().to(device), (1, 1, 1, 1))
mri_in = img_tensor[:, :, :-1, :-1]
mri = MRI(mask, device)
mri.set_h(0.5)
y = mri.H(mri_in)


mrs = MultiResSolver('fm', 'tv_l2', lmbda =9.206e-6, mask=mask, h_init=8, range_r = 512, N_scales=6, device=device, verbose=True, toi=1e-6)
mrs.solve_fm(y)
```

```python
torch.save(mrs.sols[0].to('cpu'), 'ExpB/saved_results/mrs_tv0.pt')
torch.save(mrs.sols[1].to('cpu'), 'ExpB/saved_results/mrs_tv1.pt')
torch.save(mrs.sols[2].to('cpu'), 'ExpB/saved_results/mrs_tv2.pt')
torch.save(mrs.sols[3].to('cpu'), 'ExpB/saved_results/mrs_tv3.pt')
torch.save(mrs.sols[4].to('cpu'), 'ExpB/saved_results/mrs_tv4.pt')
torch.save(mrs.sols[5].to('cpu'), 'ExpB/saved_results/mrs_tv5.pt')

torch.save(mrs.costs, 'ExpB/saved_results/mrs_tv_cost.pt')
torch.save(mrs.losses, 'ExpB/saved_results/mrs_tv_loss.pt')
torch.save(mrs.reg_values, 'ExpB/saved_results/mrs_tv_reg.pt')
torch.save(mrs.mses, 'ExpB/saved_results/mrs_tv_mse.pt')
torch.save(mrs.iters, 'ExpB/saved_results/mrs_tv_iters.pt')
torch.save(img_tensor, 'ExpB/saved_results/mrs_tv_img.pt')
torch.save(y, 'ExpB/saved_results/mrs_tv_y.pt')
```
